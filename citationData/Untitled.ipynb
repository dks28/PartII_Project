```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import beta as beta
```

```python
data = open('DBLP-citation-Feb21.txt')
line = data.readline()
citations = []
cits=0
while(line != None and cits < 2500000):
    nc = len('#citation')
    if line[:nc] == '#citation':
        citations.append(int(line[nc:]))
        cits += 1
    line = data.readline()
```

```python
citations = list(map(lambda x: max(x, 0), citations))
citations = np.random.choice(citations, 100000, replace=False)
len(citations)
```

```python
maxC = max(citations)
n, k, _ = plt.hist(citations, bins=maxC)
proportion_in_degree = n/len(citations)
cum_prop_in_degree = [sum(proportion_in_degree[i:]) for i in range(len(n))]
```

```python
cum_prop_in_degree[:10]
```

```python
plt.scatter(x=k[:-1],y=cum_prop_in_degree)
plt.yscale('log')
plt.xscale('log')
plt.ylim(1/100000, 1)
plt.xlim(1,100000)
plt.plot(k[1:]-1, (beta(k[1:]-1+2.8486, (16.323+2.8486)/16.323)/beta(2.8486, (16.323+2.8486)/16.323)), c='orange', linewidth=1, label='Analytical Solution to Price\'s Classical Model')
```
